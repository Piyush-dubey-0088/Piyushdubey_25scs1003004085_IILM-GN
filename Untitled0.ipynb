```python
# ------------------------------------------------------------
# HUMAN DISEASE PREDICTION USING SYMPTOMS
# Extended, Full ML Project (Google Colab Ready)
# Dataset AUTO-GENERATED by AI
# Model used: Multinomial Naive Bayes
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import random

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import joblib


# ------------------------------------------------------------
# STEP 1: AUTO GENERATE DATASET
# ------------------------------------------------------------

def generate_dataset():

    symptoms = [
        "itching", "skin_rash", "continuous_sneezing", "shivering", "chills",
        "joint_pain", "stomach_pain", "acidity", "vomiting", "fatigue",
        "weight_loss", "anxiety", "cold_hands", "cough", "chest_pain",
        "fever", "sweating", "dehydration", "headache", "nausea",
        "blurred_vision", "dizziness", "muscle_pain", "constipation",
        "diarrhea", "sore_throat", "shortness_of_breath", "yellow_skin",
        "blood_in_urine", "painful_urination", "back_pain", "abdominal_pain",
        "loss_of_appetite", "body_ache", "red_eyes", "fast_heart_rate",
        "swelling", "weakness", "difficulty_sleeping"
    ]

    diseases = [
        "Allergy", "Common Cold", "Flu", "Migraine", "Diabetes",
        "Hypertension", "Food Poisoning", "Malaria", "Dengue", "Jaundice",
        "Pneumonia", "Kidney Infection", "Arthritis", "Heart Disease",
        "Anxiety Disorder"
    ]

    rows = 1500
    dataset = []

    for _ in range(rows):
        symptom_values = [random.randint(0, 1) for _ in range(len(symptoms))]

        disease = random.choice(diseases)

        dataset.append(symptom_values + [disease])

    columns = symptoms + ["Disease"]
    df = pd.DataFrame(dataset, columns=columns)

    df.to_csv("symptoms_disease_dataset.csv", index=False)
    print("Dataset generated and saved as symptoms_disease_dataset.csv\n")

    return df


# ------------------------------------------------------------
# STEP 2: PREPROCESSING
# ------------------------------------------------------------

def preprocess_data(df):

    print("Preprocessing data...\n")

    X = df.drop("Disease", axis=1)
    y = df["Disease"]

    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    print("Encoding Complete!")
    return X, y, label_encoder


# ------------------------------------------------------------
# STEP 3: TRAIN MODEL
# ------------------------------------------------------------

def train_model(X, y):

    print("\nTraining Model...\n")

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42
    )

    model = MultinomialNB()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

    print("\nModel Training Completed!\n")
    return model


# ------------------------------------------------------------
# STEP 4: SAVE MODEL
# ------------------------------------------------------------

def save_files(model, encoder):
    joblib.dump(model, "disease_model.pkl")
    joblib.dump(encoder, "label_encoder.pkl")
    print("Model and encoder saved successfully!\n")


# ------------------------------------------------------------
# STEP 5: PREDICT
# ------------------------------------------------------------

def predict_disease(model, encoder, sample):
    sample_arr = np.array(sample).reshape(1, -1)
    pred = model.predict(sample_arr)
    return encoder.inverse_transform(pred)[0]


# ------------------------------------------------------------
# MAIN PIPELINE
# ------------------------------------------------------------

def main_flow():
    df = generate_dataset()
    X, y, encoder = preprocess_data(df)
    model = train_model(X, y)
    save_files(model, encoder)
    return model, encoder, df.columns.tolist()


# ------------------------------------------------------------
# RUN FULL PIPELINE
# ------------------------------------------------------------

model, encoder, cols = main_flow()

# ------------------------------------------------------------
# SAMPLE PREDICTION
# ------------------------------------------------------------
print("\n--- SAMPLE PREDICTION ---")
sample = [1] * (len(cols) - 1)   # turn all symptoms ON
print("Predicted Disease:", predict_disease(model, encoder, sample))
```
